```python
# ============================================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# ============================================================

# PyTorch para redes neuronales
import torch

# Dataset y DataLoader
from torch.utils.data import Dataset, DataLoader

# Manipulación de datos
import numpy as np
import pandas as pd

# Visualización
from matplotlib import pyplot as plt

# Barra de progreso
from tqdm import tqdm

print("Librerías cargadas correctamente.")
```

```python
# ============================================================
# 2. DETECTAR GPU
# ============================================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("Dispositivo utilizado:", device)
```

```python
# ============================================================
# 3. CARGAR DATASET
# ============================================================

columnas = [
'A1','A2','A3','A4','A5','A6','A7','A8',
'A9','A10','A11','A12','A13','A14','A15','A16'
]

data = pd.read_csv(
    "crx.data",
    header=None,
    names=columnas
)

print("Dataset cargado correctamente.")
print("Dimensiones:", data.shape)

display(data.head())
```

```python
# ============================================================
# 4. EXPLORACIÓN INICIAL
# ============================================================

print("Columnas del dataset:")
print(list(data.columns))

print("\nTipos de datos:")
print(data.dtypes)

print("\nValores nulos:")
print(data.isnull().sum())

print("\nCantidad de '?' por columna:")
print((data == '?').sum())
```

```python
# ============================================================
# 5. REEMPLAZAR '?' POR NaN
# ============================================================

data = data.replace('?', np.nan)

print("Valores '?' convertidos a NaN.")

print("\nNulos por columna:")
print(data.isnull().sum())
```

```python
# ============================================================
# 6. CONVERTIR COLUMNAS NUMÉRICAS
# ============================================================

for col in ['A2','A3','A8','A11','A14','A15']:
    
    data[col] = pd.to_numeric(data[col], errors='coerce')

print("Columnas numéricas convertidas.")
```

```python
# ============================================================
# 7. IMPUTAR VALORES NULOS
# ============================================================

# Columnas numéricas del dataset
numericas = ['A2','A3','A8','A11','A14','A15']

# Columnas categóricas
categoricas = [
'A1','A4','A5','A6','A7',
'A9','A10','A12','A13'
]

# Imputar columnas numéricas con la MEDIA
for col in numericas:

    data[col] = data[col].fillna(data[col].mean())

# Imputar columnas categóricas con la MODA
for col in categoricas:

    data[col] = data[col].fillna(data[col].mode()[0])

print("Valores nulos imputados correctamente.")

print("\nNulos restantes por columna:")
print(data.isnull().sum())
```

```python
# ============================================================
# 8. CONVERTIR TARGET A BINARIO
# ============================================================

data['A16'] = data['A16'].map({
'+':1,
'-':0
})

print("Distribución de clases:")
print(data['A16'].value_counts())
```

```python
# ============================================================
# 9. SELECCIÓN DE VARIABLES
# ============================================================

features = data.columns[:-1]

target = 'A16'

df = data.copy()

print("Dimensiones del dataset final:", df.shape)

display(df.head())
```

```python
# ============================================================
# 10. PREPARAR MATRICES X e y
# ============================================================

X = df[features].values

y = df[target].values

print("Forma de X:", X.shape)
print("Forma de y:", y.shape)
```
