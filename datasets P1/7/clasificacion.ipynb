```python
# ==========================================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# ==========================================================

# PyTorch
import torch

# Dataset y DataLoader
from torch.utils.data import Dataset, DataLoader

# Manipulación de datos
import numpy as np
import pandas as pd

# Gráficos
from matplotlib import pyplot as plt

# Barra de progreso
from tqdm import tqdm

print("Librerías cargadas correctamente.")
```

```python
# ==========================================================
# 2. DETECTAR GPU
# ==========================================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("Dispositivo utilizado:", device)
```

```python
# ==========================================================
# 3. CARGAR DATASET
# ==========================================================

data = pd.read_csv(
    "australian.dat",
    header=None,
    sep=r"\s+"
)

print("Dataset cargado correctamente.")

print("Dimensiones del dataset:")
print(data.shape)

print("\nPrimeras filas:")
print(data.head())
```

```python
# ==========================================================
# 4. ASIGNAR NOMBRES A COLUMNAS
# ==========================================================

data.columns = [
"A1","A2","A3","A4","A5",
"A6","A7","A8","A9","A10",
"A11","A12","A13","A14",
"A15"
]

print("Columnas asignadas correctamente.")
print(data.columns)
```

```python
# ==========================================================
# 5. EXPLORACIÓN INICIAL
# ==========================================================

print("Primeras filas del dataset:")
display(data.head())

print("\nTipos de datos:")
print(data.dtypes)

print("\nEstadísticas básicas:")
display(data.describe())
```

```python
# ==========================================================
# 6. CONVERTIR "?" EN NaN
# ==========================================================

data = data.replace("?", np.nan)

print("Valores '?' convertidos a NaN")

print("\nValores nulos por columna:")
print(data.isnull().sum())
```

```python
# ==========================================================
# 7. CONVERTIR COLUMNAS NUMÉRICAS
# ==========================================================

numeric_cols = ["A2","A3","A8","A11","A14"]

for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors="coerce")

print("Columnas numéricas convertidas.")
print(data.dtypes)
```

```python
# ==========================================================
# 8. IMPUTAR VALORES NULOS
# ==========================================================

for col in data.columns:

    if data[col].dtype == "object":
        data[col] = data[col].fillna(data[col].mode()[0])

    else:
        data[col] = data[col].fillna(data[col].mean())

print("Valores nulos imputados.")

print("\nNulos restantes:")
print(data.isnull().sum())
```

```python
# ==========================================================
# 9. SEPARAR VARIABLES Y TARGET
# ==========================================================

features = [
"A1","A2","A3","A4","A5",
"A6","A7","A8","A9","A10",
"A11","A12","A13","A14"
]

target = "A15"

X = data[features].values
y = data[target].values

print("Forma de X:", X.shape)
print("Forma de y:", y.shape)
```

```python
# ==========================================================
# 10. NORMALIZACIÓN
# ==========================================================

def normalizar(X):

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)

    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma


X_norm, mu, sigma = normalizar(X)

print("Datos normalizados correctamente.")
```

```python
# ==========================================================
# 11. TRAIN / TEST SPLIT
# ==========================================================

split = int(len(X_norm) * 0.8)

X_train = X_norm[:split]
y_train = y[:split]

X_test = X_norm[split:]
y_test = y[split:]

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```
