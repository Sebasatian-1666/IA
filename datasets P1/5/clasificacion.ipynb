```python
# ============================================================
# 1. IMPORTACIÓN DE LIBRERÍAS
# ============================================================

# PyTorch para redes neuronales
import torch

# Dataset y DataLoader
from torch.utils.data import Dataset, DataLoader

# Manipulación de datos
import numpy as np
import pandas as pd

# Visualización
from matplotlib import pyplot as plt

# Barra de progreso
from tqdm import tqdm

print("Librerías cargadas correctamente.")
```

```python
# ============================================================
# 2. DETECTAR GPU
# ============================================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("Dispositivo utilizado:", device)
```

```python
# ============================================================
# 3. CARGAR DATASET
# ============================================================

columnas = [
'age','workclass','fnlwgt','education','education-num',
'marital-status','occupation','relationship','race',
'sex','capital-gain','capital-loss','hours-per-week',
'native-country','income'
]

data = pd.read_csv(
    "adult.data",
    header=None,
    names=columnas,
    skipinitialspace=True
)

print("Dataset cargado correctamente.")
print("Dimensiones:", data.shape)

display(data.head())
```

```python
# ============================================================
# 4. EXPLORACIÓN INICIAL
# ============================================================

print("Columnas del dataset:")
print(list(data.columns))

print("\nTipos de datos:")
print(data.dtypes)

print("\nValores nulos:")
print(data.isnull().sum())

print("\nValores '?' por columna:")
print((data == '?').sum())
```

```python
# ============================================================
# 5. REEMPLAZAR VALORES '?' POR NaN
# ============================================================

data = data.replace('?', np.nan)

print("Valores '?' convertidos a NaN")

print("\nNulos por columna:")
print(data.isnull().sum())
```

```python
# ============================================================
# 6. ELIMINAR COLUMNAS NO NECESARIAS
# ============================================================

data = data.drop(columns=['education','fnlwgt'])

print("Columnas restantes:")
print(list(data.columns))
```

```python
# ============================================================
# 7. IMPUTAR VALORES NULOS
# ============================================================

for col in ['workclass','occupation','native-country']:

    data[col] = data[col].fillna(data[col].mode()[0])

print("Nulos imputados con moda.")

print("\nNulos restantes:")
print(data.isnull().sum())
```

```python
# ============================================================
# 8. CONVERTIR TARGET A BINARIO
# ============================================================

data['income'] = data['income'].apply(
    lambda x: 1 if x == '>50K' else 0
)

print("Variable objetivo convertida.")
print(data['income'].value_counts())
```

```python
# ============================================================
# 9. SELECCIÓN DE FEATURES
# ============================================================

features = [
'age',
'workclass',
'education-num',
'marital-status',
'occupation',
'relationship',
'race',
'sex',
'capital-gain',
'capital-loss',
'hours-per-week',
'native-country'
]

target = 'income'

df = data[features + [target]].copy()

print("Dimensiones:", df.shape)

display(df.head())
```

```python
# ============================================================
# 10. PREPARAR X e y
# ============================================================

X = df[features].values
y = df[target].values

print("Forma de X:", X.shape)
print("Forma de y:", y.shape)
```
